# Imported relevant libraries

```python
import numpy as np
import pandas as pd

import tensorflow as tf
```

# Preprocessing the Data

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe( include='all' )
```

```python
data['scaleOfIncident'] = data['scaleOfIncident'].map({'low':0, 'medium':1, 'high':2, 'catastrophic':3})
```

```python
data
```

```python
data = data.drop(['Id','incidentCode','incidentID'], axis = 1)
data
```

# Scaling the Data

```python
target = data['scaleOfIncident']

from sklearn.preprocessing import MinMaxScaler as mms

scaler = mms()
x_scaled = scaler.fit_transform(data.drop(['scaleOfIncident'], axis = 1))
x_scaled
```

# Splitting Data into train and validation sets

```python
from sklearn.model_selection import train_test_split as tts

x_train, x_vld, y_train, y_vld = tts(x_scaled, target, test_size = 0.2, random_state = 42)
```

# DNN

```python
dnn = tf.keras.models.Sequential() 
dnn.add(tf.keras.layers.Dense(units = 200, activation = 'relu'))
dnn.add(tf.keras.layers.Dense(units = 200, activation = 'relu'))
dnn.add(tf.keras.layers.Dense(units = 200, activation = 'relu'))
dnn.add(tf.keras.layers.Dense(units = 200, activation = 'relu'))
dnn.add(tf.keras.layers.Dense(units = 200, activation = 'relu'))
dnn.add(tf.keras.layers.Dense(units = 4, activation = 'softmax'))

dnn.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

dnn.fit(x_train, y_train, batch_size = 32, epochs = 20, validation_data = (x_vld, y_vld))
```

```python
data_test = pd.read_csv('test.csv')
data_test
```

```python
data_test = data_test.drop(['Id','incidentCode','incidentID'], axis = 1)

x_test_scaled = scaler.transform(data_test)
x_test_scaled
```

```python
predictions = dnn.predict(x_test_scaled, verbose=2)
predictions
```

```python
pred = []
legend = {0:'low', 1:'medium', 2:'high', 3:'catastrophic'}
for i in predictions:
    maximum = max(i)
    for j in range(4):
        if i[j] == maximum:
            pred.append(legend[j])
print(pred)
```

# Exporting data into CSV file

```python
file = open("final.csv", 'w')
file.write('Id,scaleOfIncident\n')    
for i in range(1000):
    num = 9000 + i
    file.write(str(num) + ',' + str(pred[i]) + '\n')  
    print(num)
```
